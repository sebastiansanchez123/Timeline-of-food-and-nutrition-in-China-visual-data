```python
from matplotlib import pyplot as plt
```

```python
# References: https://books.google.com.ar/books?id=guo9DwAAQBAJ&pg=PA251&lpg=PA251&dq=Sichuan+famine,+1941&source=bl&ots=ZJv-8Wve9P&sig=ACfU3U3T4SYFC13fZPL-pMlnLIkOmlOCBA&hl=en&sa=X&ved=2ahUKEwi8yPiQ95_pAhUpD7kGHcV7C0YQ6AEwCnoECAcQAQ#v=onepage&q=Sichuan%20famine%2C%201941&f=false
year = [1958, 1959, 1960, 1961, 1962, 1963, 1964, 1965]
output = [200, 170, 143.5, 147.5, 160, 170, 187.5, 194.5]
procurements = [41.8, 47.6, 30.85, 25.81, 25.76, 28.9, 31.88, 33.65]
net_rural_availability_output_minus_procurements = [158.2, 122.4, 112.7, 121.7, 134.2, 141.1, 155.6, 160.9]
```

```python
plt.figure(figsize=(15,9))
plt.plot(year, output)
plt.plot(year, procurements)
plt.plot(year, net_rural_availability_output_minus_procurements)
plt.xlabel('Year')
plt.ylabel('China grain output and state procurements, in million tons')
plt.legend(['Output', 'Procurements', 'Net rural availability (output minus procurements)'])
plt.show()
```
